```python
import pandas as pd
import time
import re
import tweepy
import requests
from bs4 import BeautifulSoup
from types import MappingProxyType
import datetime
import math

### 定数定義
```

```python
df = pd.DataFrame()

# 1秒スリープ
time.sleep(1)

# スクレイピング対象のURLを生成する
url = 'https://www.spotrac.com/epl/manchester-city-fc/contracts/'
df = pd.read_html(url)[0]
# マルチインデックスを解消する
df = df.T.reset_index(level=0, drop=True).T

html = requests.get(url)
html.encoding = "EUC-JP"
# HTMLを解析する
soup = BeautifulSoup(html.text, "lxml")

# 選手の契約情報を取得
player_name_list = []
avg_salary_list = []
expired_year_list = []

for num in range(len(df)):
    # 選手
    player_name_list.append(df[0][num])
    
    # 年俸
    # £:スターリング・ポンド⇒円変換(1£=160円)
    # https://yuis-programming.com/?p=511　※moduleをレイヤー追加するのが手間
    if df[4][num] == '-':
        df[4][num] = '£0,0'

    salary = df[4][num].replace('£', '').replace(',', '')
    salary_yen = 160 * int(salary)
    avg_salary_list.append(salary_yen)
    
    # 契約期間
    expired_year_list.append(df[6][num])
    
print(player_name_list, avg_salary_list, expired_year_list)
```
